```python
import ast
def get_name(lst):
  lst1 = []
  for i in ast.literal_eval(lst):
    lst1.append(i["name"])

  return lst1
```

```python
movies["keywords"] = movies["keywords"].apply(get_name)
```

```python
movies.head()
```

```python
import ast
def fetch_charecters(lst):
  lst1 = []
  for i in ast.literal_eval(lst):
    lst1.append(i["character"])

  return lst1
```

```python
movies["cast"] = movies["cast"].apply(fetch_charecters)
```

```python
movies.head()
```

```python
import ast
def fetch_director(lst):
  lst1 = []
  for i in ast.literal_eval(lst):
    if(i["job"] == "Director"):
      lst1.append(i["name"])
      break
  return (lst1)
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
# def split1(lst):
#   lst2 = []
#   for i in lst:
#     i = i.split()
#     lst2.append(i)
#   return lst2
# lst = ["Hi how are you my name is karting"]
# split1(lst)

movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Now we have to remove spaces in between as kartik tiwari & kartik rastogi
#Now tags for both will be kartik as common now this will make our model less accurate

# def rem_spaces(lst):
#   lst1 = []
#   for i in lst:
#     i = i.replace(" " , "")
#     lst1.append(i)
#   return lst1

# rem_spaces(["Ac tion", "Adven  ture", "Fa ntasy", "Scie nceF ict ion"])

movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" " , "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" " , "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" " , "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" " , "") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head(1)
```

```python
new_df = movies[["movie_id" , "title" , "tags"]]
new_df.head()
```

```python
#now we will convert out tags list to string

new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
# We have converted back to lower case all the charecters present at
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df["tags"][0]
```

##Now we are going to train our dataset

```python
# stop words are words which are used for sentence formation but does not contribute to

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = "english" , max_features = 5000 )
cv.fit_transform(new_df["tags"]).toarray().shape
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
port = PorterStemmer()
def stemming(text):
  lst = []
  for i in text.split():
    lst.append(port.stem(i))
  return " ".join(lst)
```

```python
new_df["tags"] = new_df["tags"].apply(stemming)
```

```python
cv = CountVectorizer(stop_words = "english" , max_features = 5000 )
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[2])),reverse = True , key = lambda x:x[:1])
```

```python
def recommender_system(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[0:6]
  for i in movie_list:
    print(new_df.iloc[i[0]].title , end = " , ")
recommender_system("Avatar")
```
